```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/creditcard.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(how ="any",inplace = True)
df['Class'] = df['Class'].astype(int)
```

```python
df.isnull().sum()
```

```python
#distribution of legit and fraud transaction
df['Class'].value_counts()
```

```python
# 0 -> noramal transaction
# 1-> falut trasacrion
legit = df[df.Class == 0]
fraud = df[df.Class==1]
```

```python
legit.shape , fraud.shape
```

```python
#statistical measures
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#undersampling
#build a sample dataset containg simiral distribution and fraud transaction
legit_sample = legit.sample(n=500)
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
#spliting
X = new_dataset.drop(columns = "Class",axis = 1)
Y  = new_dataset["Class"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33,stratify=Y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model =  LogisticRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_train)
```

```python
accuray_ = accuracy_score(y_pred,y_train)
```

```python
accuray_
```

```python
#accuracy on test data
X_test_pred = model.predict(X_test)
data_accuracy = accuracy_score(X_test_pred,y_test)
```
